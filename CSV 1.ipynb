```python
 import pandas as pd
from sqlalchemy import create_engine
```

```python
#Extract CSVs into DataFrames

crime_file = "C:/Users/Owner/desktop/ETL-Project/Resources/crime_by_state.csv"
crime_df = pd.read_csv(crime_file)
crime_df
```

```python
#Create a filtered dataframe from specific columns
crime_cols = ["STATE/UT", "Year", "Murder", "Assault on women", "Kidnapping and Abduction", "Dacoity", "Robbery", "Arson", "Hurt", "Prevention of atrocities (POA) Act", "Protection of Civil Rights (PCR) Act", "Other Crimes Against SCs"]
crime_transformed= crime_df[crime_cols].copy()
```

```python
#Rename the column headers
crime_transformed = crime_transformed.rename(columns={"State": "STATE/UT",
                                                    "Year (Count)": "Year",
                                                    "Murder (Count)": "Murder",
                                                    "Assault on Women": "Assault on women",
                                                    "Kidnapping And Abduction": "Kidnapping and Abduction",
                                                    "Dacoity (Count)": "Dacoity",
                                                    "Robbery (#)":"Robbery",
                                                    "Arson (#)": "Arson",
                                                    "Hurt (#)": "Hurt",
                                                    "POA Act": "Prevention of atrocities (POA) Act",
                                                    "PCR Act": "Protection of Civil Rights (PCR) Act",
                                                    "Other Crimes Against SC's": "Other Crimes Against SCs"})

#Clean the data by dropping duplicates setting the index
crime_transformed.drop_duplicates("STATE/UT", inplace=True)
crime_transformed.set_index("STATE/UT", inplace=True)

crime_transformed.head()
```

```python
#Create database connection
connection_string = "postgres:postgres@localhost:5432/crime_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#confirm tables
engine.tables_names()
```

```python
#Load Dataframes into Database
crime_transformed.to_sql(name='crime', con=engine, if_exists='append', index=True)
```
