```python
 import pandas as pd
from sqlalchemy import create_engine
```

```python
#Extract CSVs into DataFrames

shopping_file = "C:/Users/Owner/desktop/ETL-Project/Resources/etsy_shops_data.csv"
shopping_df = pd.read_csv(shopping_file)
shopping_df.head()
```

```python
#Create a filtered dataframe from specific columns
shopping_cols = ["creation_date", "listing_active_count", "num_favorers", "currency_code", "is_shop_us_based", "sale_message", "sales_count", "review_count", "shop_location"]
shopping_transformed = shopping_df[shopping_cols].copy()
```

```python
#Rename the column Headers 
shopping_transformed = shopping_transformed.rename(columns={"creation_date": "Creation Date",
                                                             "listing_active_count": "Listing Active Count",
                                                             "num_favorers": "Num Favorers",
                                                             "currency_code": "Currency Code",
                                                             "is_shop_us_based": "Is Shop Us Based",
                                                             "sale_message": "Sale Message",
                                                             "sales_count": "Sales Count",
                                                             "review_count": "Review Count",
                                                             "shop_location": "Shop Location"})

#Clean the Data by dropping duplicates and setting the index
shopping_transformed.set_index("Creation Date", inplace=True)

shopping_transformed.head()
```

```python
#Create Database Connection
connection_string = "postgres:postgres@localhost:5432/shopping_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#Confirm Tables
engine.tables_names()
```

```python
#Load Dataframes into Database
shopping_transformed.to_sql(name='shopping', con=engine, if_exists='append', index=True)
```
